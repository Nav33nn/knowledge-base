```python
cleaned_df = df.dropna(axis=1,how='all')
```

```python
cleaned_df.to_csv('../data/cleaned_loan.csv', index=False)
```

```python
cols = list(cleaned_df)
nunique = cleaned_df.apply(pd.Series.nunique
                          )
print(nunique)
cols_to_drop = nunique[nunique == 1].index
#df.drop(cols_to_drop, axis=1)
print(cols_to_drop)
```

```python
cleaned_df = cleaned_df.drop(columns=cols_to_drop)
```

```python
cleaned_df.shape
```

```python
cleaned_df.to_csv('../data/cleaned_loan_001.csv', index=False)
```

```python
cleaned_df['term']=cleaned_df['term'].apply(lambda x:x.split()[0])
cleaned_df['int_rate']=cleaned_df['int_rate'].apply(lambda x:x.split('%')[0])
cleaned_df['emp_length'] = cleaned_df['emp_length'].apply(lambda x:str(x).replace('< 1','0').replace('10+','10').split()[0])
#cleaned_df['emp_length']=cleaned_df['emp_length'].apply(lambda x:x if len(str(x))<2 else x.replace('< 1','0').replace('10+','10'))
#cleaned_df['emp_length']=cleaned_df['emp_length'].apply(lambda x:x.split()[0])
```
